```python
import pandas as pd
import re
import numpy as np
```

```python
df = pd.read_csv('../data/all_offers_infos_raw.csv')
```

```python
rent_list = []
nbk_list = []
rooms_list = []
area_list = []
byr_list = []
energy_list = []
links_list = []
for offer in range(len(df)):
    # separate info by <div>
    individual_list = df['html_infos'][offer].replace('\t', '').split('<div')
    
    # get rent price
    try: rent_price = [re.findall('\d+.0', x) for x in individual_list if 'priceid_2' in x][0][1]
    except: rent_price = np.nan
    rent_list.append(rent_price)
    
    # extra montly costs
    try: nbk = [re.findall('\d+.0', x) for x in individual_list if 'priceid_2' in x][1][1]
    except: nbk = np.nan
    nbk_list.append(nbk)
    
    # rooms
    try: rooms = [re.findall('\d+', x) for x in individual_list if 'equipmentid_1' in x][0][-2]
    except: rooms = np.nan
    rooms_list.append(rooms)
    
    # area
    try: area = [re.findall('\d+', x) for x in individual_list if 'areaid_1' in x][0][-2]
    except: area = np.nan
    area_list.append(area)
    
    # building year
    try: byr = [re.findall('\d+', x) for x in individual_list if 'yearbuild' in x][0][-1]
    except: byr = np.nan
    byr_list.append(byr)
    
    # energy class
    try: energy = [re.findall('>\w+ \w', x) for x in individual_list if 'efficiencyValue' in x][0][0].replace('>', '')
    except: energy = np.nan
    energy_list.append(energy)
    
    # offer link
    try: href = [re.findall('\w+://\w+.\w+.\w+/\w+/\d+', x) for x in individual_list if 'www' in x][0][0]
    except: href = np.nan
    links_list.append(href)
```
